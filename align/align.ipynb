# Align

Align two pdbs

reference: https://github.com/charnley/rmsd

```python
p1 = './SeUGI-source.pdb'
p2 = './SeUGI-target.pdb'
p3 = './SeUGI-result.pdb'
```

```python
import sys
sys.path.insert(0, '../')
from atom import Protein, RMSD
```

```python
proteinA, proteinB = None, None
with open(p1) as f:
    proteinA = Protein.fromFile(f)
    proteinA.centroid()

with open(p2) as f:
    proteinB = Protein.fromFile(f)
    proteinB.centroid()

P = proteinA.toNumpy()
Q = proteinB.toNumpy()

with open(p3, 'w') as f:
    P2 = RMSD.kabsch_rotate(P, Q)
    proteinA.fromNumpy(P2)
    f.write(proteinA.output())
```


+ Green: SeUGI_source.pdb which rotates SeUGI_target by a random angle.
+ Blue: SeUGI_target.pdb
+ Purple: SeUGI_result.pdb which is generated by using Kabsch algorithm
